```python
datetime_dim = df[['tpep_pickup_datetime','tpep_dropoff_datetime']].drop_duplicates().reset_index(drop=True)
```

```python
datetime_dim.head()
```

```python
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime']=pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
datetime_dim['tpep_pickup_datetime']=pd.to_datetime(datetime_dim['tpep_pickup_datetime'])
datetime_dim['tpep_dropoff_datetime']=pd.to_datetime(datetime_dim['tpep_dropoff_datetime'])
datetime_dim['pick_day']=datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['pick_month']=datetime_dim['tpep_pickup_datetime'].dt.month
datetime_dim['pick_year']=datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['pick_weekday']=datetime_dim['tpep_pickup_datetime'].dt.weekday
datetime_dim['pick_hour']=datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['drop_day']=datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['drop_month']=datetime_dim['tpep_dropoff_datetime'].dt.month
datetime_dim['drop_year']=datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['drop_weekday']=datetime_dim['tpep_dropoff_datetime'].dt.weekday
datetime_dim['drop_hour']=datetime_dim['tpep_dropoff_datetime'].dt.hour
```

```python
datetime_dim.drop('datetime_id',axis=1,inplace=True)
datetime_dim.insert(0,'datetime_id',datetime_dim.index)
```

```python
datetime_dim.head()
```

```python
pickup_location_dim=df[['pickup_longitude','pickup_latitude']].drop_duplicates().reset_index(drop=True)
pickup_location_dim.insert(0,'pickup_location_id',pickup_location_dim.index)
pickup_location_dim.head()
```

```python
dropoff_location_dim=df[['dropoff_longitude','dropoff_latitude']].drop_duplicates().reset_index(drop=True)
dropoff_location_dim.insert(0,'dropoff_location_id',dropoff_location_dim.index)
dropoff_location_dim.head()
```

```python
passenger_count_dim=df[['passenger_count']].drop_duplicates().reset_index(drop=True)
passenger_count_dim.insert(0,'passenger_count_id',passenger_count_dim.index)
passenger_count_dim.head()
```

```python
trip_distance_dim=df[['trip_distance']].drop_duplicates().reset_index(drop=True)
trip_distance_dim.insert(0,'trip_distance_id',trip_distance_dim.index)
trip_distance_dim.head()
```

```python
rate_code_dim=df[['RatecodeID']].drop_duplicates().reset_index(drop=True)
rate_code_dim.insert(0,'rate_code_id',rate_code_dim.index)
rate_code_dim.head()
```

```python
store_and_fwd_flag_dim=df[['store_and_fwd_flag']].drop_duplicates().reset_index(drop=True)
store_and_fwd_flag_dim.insert(0,'store_and_fwd_flag_id',store_and_fwd_flag_dim.index)
store_and_fwd_flag_dim.head()
```

```python
payment_type_dim=df[['payment_type']].drop_duplicates().reset_index(drop=True)
payment_type_dim.insert(0,'payment_type_id',payment_type_dim.index)
payment_type_dim.head()
```

```python
fact_table = df.merge(datetime_dim, on=['tpep_pickup_datetime', 'tpep_dropoff_datetime']) \
              .merge(passenger_count_dim, on='passenger_count') \
              .merge(trip_distance_dim, on='trip_distance') \
              .merge(rate_code_dim, on='RatecodeID') \
              .merge(pickup_location_dim, on=['pickup_longitude', 'pickup_latitude']) \
              .merge(dropoff_location_dim, on=['dropoff_longitude', 'dropoff_latitude']) \
              .merge(payment_type_dim, on='payment_type') \
              .merge(store_and_fwd_flag_dim, on='store_and_fwd_flag')
```

```python
fact_table=fact_table[['VendorID','datetime_id','passenger_count_id','trip_distance_id','pickup_location_id','dropoff_location_id','rate_code_id','store_and_fwd_flag_id','payment_type_id','fare_amount','extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount']]
```

```python
fact_table.head()
```

```python
datetime_dim.to_csv('./datetime_dim.csv',index=False)
pickup_location_dim.to_csv('./pickup_location_dim.csv',index=False)
dropoff_location_dim.to_csv('./dropoff_location_dim.csv',index=False)
passenger_count_dim.to_csv('./passenger_count_dim.csv',index=False)
trip_distance_dim.to_csv('./trip_distance_dim.csv',index=False)
rate_code_dim.to_csv('./rate_code_dim.csv',index=False)
store_and_fwd_flag_dim.to_csv('./store_and_fwd_flag_dim.csv',index=False)
payment_type_dim.to_csv('./payment_type_dim.csv',index=False)
fact_table.to_csv('./fact_table.csv',index=False)
```
